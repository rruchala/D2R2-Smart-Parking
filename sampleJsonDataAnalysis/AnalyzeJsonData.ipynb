```python

from pprint import pprint
import json

with open('sample_data.json') as f:
    json_data = json.load(f)

    
pprint(json_data)

```

```python
#getting json_data into DataFrame

from pandas.io.json import json_normalize

df=json_normalize(json_data, max_level=1)
df.info()
```

```python
# Initiall look into ingested data data
df.head(10)
```

```python
import pandas as pd

df = df.astype(dtype= {
        "timestamp" : "datetime64[s]",
        "meter.address":"category",
        "meter.number":"category"})
df.info()
```
